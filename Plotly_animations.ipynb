```python
import pandas as pd
import plotly.express as px
```

```python

# Load the data
# df1 = pd.read_csv('data/car_data.csv')
df2 = pd.read_csv('data/cleaned_car_data.csv')

# Concatenate the two dataframes
# df = pd.concat([df1, df2])

df = df2

# Check for missing values
print(df.isnull().sum())

# Handle missing values (if any)
# df = df.dropna()  # or use appropriate method

# Perform EDA
print(df.describe())
```

```python
df = df.sort_values(by='Age')
```

```python
df
```

```python

fig1 = px.histogram(df, x="Price", nbins=50, title='Price Distribution')
fig1.show()
```

```python
fig2 = px.scatter(df, x="Age", y="Price", color="Fuel Type", title='Price vs Year Colored by Fuel Type')
fig2.show()
```

```python
fig3 = px.box(df, x="Fuel Type", y="Price", title='Price Distribution for Different Fuel Types')
fig3.show()
```

```python
fig = px.scatter(
    data_frame=df,
    y='Max Power',
    x='Price',
    animation_frame='Age',
    color='Fuel Type',
    size='Engine',
    log_y=True,
    hover_name="Make"
)

fig.show()
```
